```python
#histograms to find the distribution of values in single numerical varible as it 

#subplot functions allows to indicate where we have plot the graph i.e 121 whre 1 row 2 columns and 1 is where 
#to place the graph 

import numpy as np
plt.figure(figsize=(10,7))
plt.subplot(321)#(1 row 2 columns and 1 graph needs to be created)
plt.hist(x = "Attack",data = df1,bins = 10,color = "darkblue")
plt.title("Attacks")
plt.subplot(133)
plt.hist(x = "Defense",data = df1,bins=20,color = "green")#bins is used for creating the ranges or bin values and x is used to 
#show different graphs for the subplots and y is not used as it is univariate
plt.title("Defense")
plt.xticks(np.arange(0,201,10))#where we can define the x varibales how we can increament it with the number
plt.yticks(np.arange(0,50,5))#where we can define the y varibales how we can increament it with the number
plt.show()
```
