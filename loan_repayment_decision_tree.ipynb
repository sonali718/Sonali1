```python
x = df.drop(["credit.policy"],axis = 1)
```

```python
y = df["credit.policy"]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(data=ss,columns=df1.drop("credit.policy"))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=6)
```

```python
pca1 = pca.fit_transform(ss)
pca1
```

```python
df2 = pd.DataFrame(data=pca1,columns=["PC1","PC2","PC3","PC4","PC5","PC6"])
df2
```

```python
target = pd.Series(df["credit.policy"],name="credit.policy")
result = pd.concat([df2,target],axis = 1)
result
```

```python
print('Variance of each component:', pca.explained_variance_ratio_)
print('\n Total Variance Explained:', round(sum(list(pca.explained_variance_ratio_))*100, 2))
```

```python
from sklearn.model_selection import train_test_split
```

```python
df["credit.policy"].value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
os = RandomOverSampler(sampling_strategy=1)
```

```python
x_res,y_res = os.fit_sample(x,y)
x_res.shape,y_res.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size = 0.33,random_state = 10) 
```

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_test.value_counts()
```

```python
lr.score(x_test,y_test)*100
```

```python
len(df)
```

```python
y_pred = lr.predict(x_test)
y_pred
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(lr.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion="entropy",max_depth=5,max_leaf_nodes=5,random_state=50)
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```
